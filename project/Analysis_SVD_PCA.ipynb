## Aplicación de la SVD (Descomposición de Valores Singulares) al dataset

```python
df_medidas = df_triplaris.drop(['muestra'], axis=1) #Eliminamos la columna "muestra" porque no nos aporta información

print("Variables seleccionadas para el análisis:")
print(df_medidas.columns.tolist())

#Convertimos todas las columnas de objeto a float, reemplazamos comas por puntos y lo convertimos a numérico
df_medidas = (df_medidas.astype(str).replace(",", ".", regex = True).apply(pd.to_numeric, errors = "coerce"))

#ESTANDARIZACIÓN DE DATOS
#Esto lo hacemos con el objetivo de homogeneizar su escala. Debido a que tenemos a los ángulos con mayor magnitud que las
#demás mediciones, estas variables dominan el análisis. Ahora, cuando hacemos este proceso de centrar los datos nos permite
#que todas las dimensiones tengan una mejor contribución en pasos siguientes.

X = df_medidas.values #Tomamos los datos a una matriz numpy (como un array)


scaler = StandardScaler()
X_centrado = scaler.fit_transform(X)  #Ajustamos la escala: (valor - media) / desviación estándar
#Se utilizó el  étodo fit_transform() para calcular la media y la desviación estándar, y aplicar una transformación lineal
#acada variable y reescalando sus dimensiones para trabajar en un sistema de coordenadas comparables.

print("\nVerificación de Estandarización")
print(f"Media después de escalar: {np.mean(X_centrado):.2f}")
print(f"Desviación estándar: {np.std(X_centrado):.2f}")

#APLICACIÓN DE SVD (Descomposición de valores singulares)

#full_matrices=False es para la versión compacta, usada en reducción de dimensionalidad
U, S, Vt = np.linalg.svd(X_centrado, full_matrices=False)
```

```python
df_medidas = df_triplaris.drop(['muestra'], axis=1)
#Eliminamos la columna "muestra" porque no nos aporta información

print("Variables seleccionadas para el análisis:")
print(df_medidas.columns.tolist())

#Convertimos todas las columnas de objeto a float, reemplazamos comas por puntos y lo convertimos a numérico
df_medidas = (df_medidas.astype(str).replace(",", ".", regex = True).apply(pd.to_numeric, errors = "coerce"))

X = df_medidas.values #Tomamos los datos como una matriz numpy (como un array)

scaler = StandardScaler()
X_centrado = scaler.fit_transform(X)  #Ajustamos la escala: (valor - media) / desviación estándar

print("\nVerificación de Estandarización")
print(f"Media después de escalar: {np.mean(X_centrado):.2f}")
print(f"Desviación estándar: {np.std(X_centrado):.2f}")

#APLICACIÓN DE SVD
#full_matrices=False es para la versión compacta, usada en reducción de dimensionalidad
U, S, Vt = np.linalg.svd(X_centrado, full_matrices=False)
```

## Interpretación de la matriz "S"

```python

#S es el vector con valores singulares y un escalamiento diagonal, que mide cuánto se estira o se comprime la información
#en cada dimensión.

#Si el valor es alto, esa dimensión es crucial. Si es bajo, es ruido.

#Los valores singulares indican la importancia relativa de cada dimensión, permitiendo reducir la dimensión del conjunto
#sin perder información relevante.

print("VALORES SINGULARES DE CADA DIMENSIÓN")
for i, valor in enumerate(S):  #enumerate para que nos devuelva el indice
    print(f"Dimensión {i+1}: {valor:.2f}")

#Luego de analizar los valores singulares, concluimos que en la dimensión 12 es donde no hay varianza, la matriz no estira
#nada en esa dirección o que es combinación lineal de las demás. Además, la dimensión 1 es la dirección donde más datos
#varían y es la componente principal #1, es decir, que existe una dirección predominante de varianza en la morfología de la
#flor.--------------------PARA ANALSIS
```

## Interpretación de la matriz "Vt"

```python
#Vt es una matriz ortogonal cambio de base de entrada, sus columnas definen a una base ortonormal e indica desde qué
#direcciones entra la información. Nos dice qué tanto de 'radio', 'ángulo' o 'ancho' se usó para crear PC1, PC2, etc.

print(f"Vt tiene {Vt.shape[0]} filas y {Vt.shape[1]} columnas.")

print("\n")

nombres_variables = df_medidas.columns.tolist()

# Creamos un DataFrame para ver los pesos de los 3 primeros componentes
df_cargas = pd.DataFrame(Vt[:3, :], columns=nombres_variables, index=['PC1', 'PC2', 'PC3'])

print("ANÁLISIS DE VECTORES PROPIOS")
display(df_cargas.round(2))

#Valores cercanos a +1 o -1 indican que esa variable domina el componente.
#Por Ej: Si radio1 tiene 0.50, significa que el Tamaño del Radio es lo más importante de la flor.
```

## Interpretación de la matriz "U"

```python
#U es una matriz ortogonal con columna con vectores singulares izquierdos, es decir, las proyecciones de las observaciones.
#Matemáticamente, los vectores singulares izquierdos son los vectores propios de la matriz AAt.
#Sus columnas consitituyen una base ortonormal de salida y nos dice hacia donde va el vector después del escalamiento.
#En otras palabras, nos dice las coordenadas de las flores después de la transformación.

#Matemáticamente las coordenadas nuevas son U * S. Debido a que, combinamos la dirección (de U) y la magnitud (de S) que
#contiene cada dimensión

print(f"U tiene {U.shape[0]} filas y {U.shape[1]} columnas.")

Coordenadas_Nuevas = U[:, :3] * S[:3] # Tomamos solo las 3 primeras dimensiones

df_U = pd.DataFrame(Coordenadas_Nuevas, columns=['Coord_PC1', 'Coord_PC2', 'Coord_PC3'])
print("ANÁLISIS DE PROYECCIONES")
display(df_U.head().round(2))  #mostramos las 5 primeras filas
```

## Análisis de varianza por cada valor singular

```python


#Para nuestro estudio, en álgebra lineal, los valores singulares son magnitudes de deformación. Sin embargo, en estadística
#la varianza (que representa información) es proporcional al cuadrado de esos valores singulares.

eigenvalues = np.square(S) # esto equivale a los valores propios de la matriz de covarianza
varianza_total = np.sum(eigenvalues)

#calculamos qué porcentaje aporta cada dimensión
varianza_explicada = eigenvalues / varianza_total
varianza_acumulada = np.cumsum(varianza_explicada)

#Dataframe para visualizar los datos de mejor manera
componentes = [f"PC{i+1}" for i in range(len(S))]
valoresSingulares = np.round(S, 2);    #le redondeamos decimales
varianzaInd = np.round(varianza_explicada*100, 2);
varianzaAcum = np.round(varianza_acumulada*100, 2);

df_varianza = pd.DataFrame({
    "Componentes": componentes,
    "Valor Singular": valoresSingulares,
    "Varianza Individual": varianzaInd,
    "Varianza Acumulada": varianzaAcum
})

print("TABLA DE ANÁLISIS DE LA VARIANZA")
display(df_varianza)

#¿Qué es lo que podemos deducir de esta varianza?
# 1) para conservar el 90% de la información necesitamos 7 dimensiones.
# 2) con 3 dimensiones retemnemos solamente el 67% de la información.
# 3) con 2 dimensiones retenemos solamente el 54% de información.
```

## Gráficos y análisis de varianza

```python
print('Podemos observar en este gráfico que para tener una varianza (equivalente a información) de un 90% (es la más óptima)')
print('debemos de tomar 7 componentes! Un gran hallazgo de nuestro data set')
print('\n')
print('Debido a que el ser humano no puedes ver un gráfico de 7 componentes, tomaremos el gráfico de 3 componentes para tener')
print('una idea visual de lo que estamos analizando.')

#Para responder la pregunta: ¿Cuál es el mínimo de componentes que nos dan la mayor información.
#Realizaremos un gráfico scree plot mixto, en el que veremos la varianza individual y la varianza acumulada.
#Con este gráfico podremos determinar con facilidad cuánta información aporta cada componente.

# GRÁFICO 1 - Scree Plot Mixt
# Coordenadas proyectadas (U escalado por S)
PC1 = U[:, 0] * S[0]
PC2 = U[:, 1] * S[1]
PC3 = U[:, 2] * S[2]


plt.figure(figsize=(12, 4))

#barras de varianza individual
barras = plt.bar(range(1, len(S)+1), varianza_explicada, alpha=0.6, color='skyblue', label='Varianza Individual (Aporte)')

#línea de la varianza acumulada
plt.plot(range(1, len(S)+1), varianza_acumulada, marker='o', linestyle='-', color='blue', linewidth=2, label='Varianza Acumulada')

# Líneas de referencia (Umbrales)
plt.axhline(y=0.90, color='r', linestyle='--', alpha=0.8, label='Umbral 90% (Precisión)')
plt.axvline(x=7, color='r', linestyle=':', alpha=0.8) # Componente 7

# Etiquetas de texto para resaltar los puntos clave
plt.text(7.2, 0.92, 'Necesario para >90%', color='red', fontsize=10, fontweight='bold')
plt.text(3.2, varianza_acumulada[2], f'3 Comps: {varianza_acumulada[2]*100:.1f}%', color='blue', fontsize=10, fontweight='bold')

plt.xlabel('Número de Componente Principal')
plt.ylabel('Proporción de Varianza')
plt.title('Scree Plot: Decisión de Dimensionalidad (Codo vs Acumulado)')
plt.xticks(range(1, 13))
plt.legend(loc='center right')
plt.grid(True, alpha=0.3)
plt.show()


# GRÁFICO 2 - Visualización de 3 componentes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Usamos el color (c) basado en PC3 para dar sensación de profundidad
img = ax.scatter(PC1, PC2, PC3, c=PC3, cmap='viridis', s=80, edgecolor='k', alpha=0.9)

ax.set_xlabel(f'PC1 ({varianza_explicada[0]*100:.1f}%)')
ax.set_ylabel(f'PC2 ({varianza_explicada[1]*100:.1f}%)')
ax.set_zlabel(f'PC3 ({varianza_explicada[2]*100:.1f}%)')
ax.set_title(f'Proyección 3D de Triplaris (Varianza Total: {varianza_acumulada[2]*100:.1f}%)')
fig.colorbar(img, label='Valor del Componente 3')
plt.show()
```
